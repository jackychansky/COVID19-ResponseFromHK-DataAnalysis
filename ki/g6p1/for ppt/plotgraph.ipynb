```python
import pandas as pd
```

```python
df_covid=pd.read_excel('covidnew_hk.xlsx', index_col='Date')
```

```python
df_covid.index=pd.DatetimeIndex(df_covid.index)
```

```python
df_covid
```

```python
df_immi_gt = pd.read_csv('immi+googletrend.csv',index_col='Unnamed: 0')
```

```python
df_immi_gt.index=pd.DatetimeIndex(df_immi_gt.index)
```

```python
df_immi_gt
```

```python
merge=pd.merge(df_immi_gt, df_covid, how='left', left_index=True, right_index=True)
```

```python
merge
```

```python
merge.to_csv('immi_gt_covidhk.csv')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.lineplot(x = merge.index, y = "Confirmed", data = merge)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(merge.index , merge.Confirmed, color="red")
# set x-axis label
ax.set_xlabel("Date",fontsize=14)
# set y-axis label
ax.set_ylabel("Confirmed Case",color="red",fontsize=14)
# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(merge.index, merge['Mask(Chi)'] ,color="blue")
ax2.set_ylabel("Mask Google Search",color="blue",fontsize=14)
plt.show()
# save the plot as a file
fig.savefig('hk_gt_case.jpg',
            format='jpeg',
            dpi=600,
            bbox_inches='tight')
```
